```python
import talib
import pandas as pd
import pandas_ta as ta
```

```python
df = pd.read_csv("data/LTCUSDT-1h.csv")
df.set_index(pd.DatetimeIndex(df["date"]), inplace=True)
df = df.drop(
    columns=[
        "date",
        "Unnamed: 0",
        "open_time",
        "close_time",
        "quote_volume",
        "count",
        "taker_buy_volume",
        "taker_buy_quote_volume",
        "ignore",
    ]
)
df.head()
```

```python
target = [0]
for i in range(len(df) - 1):
    q1, q2 = df.iloc[i]["close"], df.iloc[i + 1]["close"]
    if q2 > q1:
        target.append(1)
    else:
        target.append(0)
df["target"] = target
```

```python
# Calculate Returns and append to the df DataFrame
df.ta.log_return(cumulative=True, append=True)
df.ta.percent_return(cumulative=True, append=True)
df.ta.sma(length=10, append=True)
df.tail()
```

```python
candles = df.ta.cdl_pattern("all", scalar=1)
```

```python
df.ta.adx(append=True)
df.ta.amat(append=True)
df.ta.aroon(append=True)
df.ta.chop(append=True)
df.ta.cksp(append=True)
df.ta.decay(append=True)
df.ta.decreasing(append=True)
df.ta.dpo(append=True)
df.ta.increasing(append=True)
df.ta.long_run(append=True)
df.ta.qstick(append=True)
df.ta.short_run(append=True)
df.ta.tsignals(append=True)
df.ta.ttm_trend(append=True)
df.ta.vhf(append=True)
df.ta.vortex(append=True)
df.ta.xsignals(append=True)

trend = df.drop(
    columns=["open", "high", "low", "close", "volume", "target", "DPO_20"]
)
```
